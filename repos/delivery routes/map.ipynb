```
import pandas as pd
import requests
from xml.etree import ElementTree
import numpy as np
import folium
```

```
df = pd.read_csv("customerdata_cleaned.csv")
df.head()
```

```
locations = df[['Latitude', 'Longitude']]
locationlist = locations.values.tolist()

map = folium.Map(location=[25.949633, -80.280748], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=df['Client'][point]).add_to(map)
map
```
